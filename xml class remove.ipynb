```python
import os
import xml.etree.ElementTree as ET
```

```python
Dir = r"C:\Users\a\Desktop\b"

file_list = os.listdir(Dir)
xml_list = []
for file in file_list:
    if '.xml' in file:
        xml_list.append(file)
```
